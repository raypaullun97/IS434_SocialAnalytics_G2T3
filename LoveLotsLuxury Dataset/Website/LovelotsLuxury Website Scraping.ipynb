```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Extraction of Handbags

```python
item_name=[]
item_picture=[]
item_brand=[]
item_price=[]
item_cat=[]
item_available=[]
```

```python
page_number = 1

while page_number <= 38:
    source_path = "https://lovelotsluxury.com/collections/frontpage?page="+ str(page_number)
    page = requests.get(source_path)
    page_content = page.content
    soup = BeautifulSoup(page_content, 'html.parser')

    all_items = soup.find_all('div', {"class": "card"})
    
    for item in all_items:
        
        image_element= item.find('div', {"class":"card__image-wrapper"})
        image_link= image_element.select_one("noscript").find('img')['src']

        name= item.find('h3', {"class": "card__name"}).text

        if item.find('span', {"class": "card__price--sale"}) != None:
            price= item.find('span', {"class": "card__price--sale"}).text
            item_available.append('Available')
            
        elif item.find('div', {"class": "card__price"}) != None:
            price= item.find('div', {"class": "card__price"}).text.strip()
            item_available.append('Available')

        
        else:
            collection_link= item.find('a')['href']
            link_reference= "https://lovelotsluxury.com/"+collection_link
            page_sold= requests.get(link_reference)
            page_content_sold= page_sold.content
            soup_sold = BeautifulSoup(page_content_sold, 'html.parser')
            price= soup_sold.find("span", {"class":"product__current-price"}).text
            item_available.append('Sold')
            
            
        brand= name.split()[0]
        item_brand.append(brand)
                
        item_name.append(name)
        item_picture.append(image_link)
        item_cat.append('Handbags')
    
      
        price = price[1:]
        x = price.split()
        i_price = x[0]
            
        if "," in i_price:
            price_raw= i_price.split(",")
            i_price= price_raw[0]+price_raw[1]
        
        item_price.append(i_price)
    
    page_number += 1
        
```
