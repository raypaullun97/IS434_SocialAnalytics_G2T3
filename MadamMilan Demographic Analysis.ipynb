```python
import pandas as pd
import os
import json
import glob
from pathlib import Path  
```

```python
# filepath = demo[1]
# df = pd.read_json(filepath, lines = True)
```

```python
json_dir = './Dataset/Milan Json_Rekognition'
json_pattern = os.path.join(json_dir, '*.json')
file_list = glob.glob(json_pattern)
file_list
```

```python
dfs = []
for file in file_list:
    with open(file) as f:
        json_data = pd.json_normalize(json.loads(f.read()))
        json_data['site'] = file.rsplit("/", 1)[-1]
    dfs.append(json_data)

df = pd.concat(dfs)
df2 = df
```

```python
df2
```

```python
#Delete Redundant Column
#df2.drop('Emotions', inplace = True, axis = 1)
df2.drop('Landmarks', inplace = True, axis = 1)
df2.drop('Confidence', inplace = True, axis = 1)
df2.drop('BoundingBox.Width', inplace = True, axis = 1)
df2.drop('BoundingBox.Height', inplace = True, axis = 1)
df2.drop('BoundingBox.Left', inplace = True, axis = 1)
df2.drop('BoundingBox.Top', inplace = True, axis = 1)
df2.drop('Smile.Value', inplace = True, axis = 1)
df2.drop('Smile.Confidence', inplace = True, axis = 1)
df2.drop('EyesOpen.Value', inplace = True, axis = 1)
df2.drop('EyesOpen.Confidence', inplace = True, axis = 1)
df2.drop('MouthOpen.Value', inplace = True, axis = 1)
df2.drop('MouthOpen.Confidence', inplace = True, axis = 1)
df2.drop('Eyeglasses.Value', inplace = True, axis = 1)
df2.drop('Eyeglasses.Confidence', inplace = True, axis = 1)
df2.drop('Sunglasses.Value', inplace = True, axis = 1)
df2.drop('Sunglasses.Confidence', inplace = True, axis = 1)
df2.drop('Beard.Value', inplace = True, axis = 1)
df2.drop('Beard.Confidence', inplace = True, axis = 1)
df2.drop('Mustache.Value', inplace = True, axis = 1)
df2.drop('Mustache.Confidence', inplace = True, axis = 1)
df2.drop('Pose.Roll', inplace = True, axis = 1)
df2.drop('Pose.Yaw', inplace = True, axis = 1)
df2.drop('Pose.Pitch', inplace = True, axis = 1)
df2.drop('Quality.Brightness', inplace = True, axis = 1)
df2.drop('Quality.Sharpness', inplace = True, axis = 1)
df2.drop('Emotions', inplace = True, axis = 1)
```

```python
df2['AverageAge'] = (df2['AgeRange.Low'] + df2['AgeRange.High']) / 2
df2.count()
```
