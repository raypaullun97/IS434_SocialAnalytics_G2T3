```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
item_name=[]
item_picture=[]
item_brand=[]
item_price=[]
item_cat=[]
item_available=[]
```

HandBags

```python
page_number = 1

while page_number <= 18:
    source_path = "https://www.madammilan.com.sg/bags.html?p="+ str(page_number)
    page = requests.get(source_path)
    page_content = page.content
    soup = BeautifulSoup(page_content, 'html.parser')

    all_items = soup.find_all('div', {"class": "item span4"})
    
    for item in all_items:
        brand= item.find('div', {"class": "brandname"}).text
        picture= item.find('img')['src']
        name= item.find('h3', {"class": "product-name"}).text
        price_data= item.find('span', {"class": "mmprice"})
        price = price_data.text[3:]
        
        if "," in price:
            price_raw= price.split(",")
            price= price_raw[0]+price_raw[1]
        
        item_name.append(name)
        item_picture.append(picture)
        item_brand.append(brand)
        item_price.append(price)
        item_available.append('Available')
        item_cat.append('Handbags')
  
    
    page_number += 1
```
