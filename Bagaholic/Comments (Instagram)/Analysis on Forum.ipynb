```python
original_post= pd.read_csv('../bag_post.csv')
original_post 
```

```python
original_post = original_post.drop_duplicates('postUrl')
original_post
```

```python
original_post['Post_Number'] = np.arange(len(original_post))
original_post['Post_Number']
```

```python
#original_post.to_csv("orginal_post_edited.csv", index = False)
```

```python
post_data= original_post[['postUrl', 'description', 'commentCount','likeCount','pubDate', 'Post_Number']]
```

```python
post_data['BreakDown']= post_data['description'].str.split()
```

```python
post_data.iloc[0][6]
```

```python
tagged_brand= []
related={}
for row in post_data.iterrows():
    breakdown=row[1]['BreakDown']
    post_no= row[1]['Post_Number'] 
    url= row[1]['postUrl']
    brand_row = []
    for word in breakdown:
        if '@' in word:
            brand = re.sub(r"[^a-zA-Z0-9]","",word)
            brand_row.append(brand)
            related[url]= post_no
            
    tagged_brand.append(brand_row)
```

```python
related
```

```python
post_data['Tagged']=tagged_brand
```

```python
post_data
```

```python
post_data.to_csv("orginal_post_edited.csv", index = False)
```

```python
cross_relate=[]
for row in comment.iterrows():
    url= row[1]['query']
    if url in related:
        post_number= related[url]
        cross_relate.append(post_number)
    else:
        cross_relate.append('error')
```

```python
comment['Post_Number']= cross_relate
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def sentiment_scores(sentence):
# Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
# polarity_scores method of SentimentIntensityAnalyzer
# oject gives a sentiment dictionary.
# which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
    negative_senti= sentiment_dict['neg']*100
    positive_senti= sentiment_dict['pos']*100
    neutral_senti= sentiment_dict['neu']*100
    
    compound_senti= sentiment_dict['compound']
    
    positive.append(positive_senti)
    negative.append(negative_senti)
    neutral.append(neutral_senti)
    compound.append(compound_senti)
```

```python
positive=[]
neutral=[]
negative=[]
compound=[]
for row in comment.iterrows():
    url= row[1]['comment']
    sentiment_scores(url)
```

```python
comment['Positive']= positive
comment['Neutral']= neutral
comment['Negative']= negative
comment['Compound']= compound
```
